```python
tags=[]
xy=[]
len_dict=0
for intent in intents:
    tag = intent['tag']
    tags.append(tag)
    for pattern in intent['patterns']:
        len_dict=len_dict + 1
        w = pattern
        xy.append((w,tag))
df = pd.DataFrame(xy)
df.rename(columns = {0:"Sentence", 1:"Target"}, inplace = True)
named_y = df['Target']
df
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
print(df['Target'])
```

```python
X_train = df.Sentence
y_train = df.Target
```

```python
labels = {}
for i in range(len_dict):
    labels[y_train[i]] = named_y[i]
labels
```

```python
vec = CountVectorizer(min_df=1)
```

```python
X_train_count = vec.fit_transform(X_train.values)
X_train_count.toarray()[0]
```

```python
bag_len = len(X_train_count.toarray()[0])
bag_len
```

```python
df_bow = pd.DataFrame(X_train_count.toarray(), columns=vec.get_feature_names_out())
df_bow
```

```python
X_train=df_bow.to_numpy()
print(X_train)
print(X_train_count)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1000, input_shape=(bag_len,),activation='relu',),
    keras.layers.Dense(100,activation='relu',),
    keras.layers.Dense(11,activation='sigmoid',)
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train, y_train, epochs=100)
```
